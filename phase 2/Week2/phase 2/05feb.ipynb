```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
##load the data set
data = pd.read_csv('car_data.csv')
data.head()
```

```python
##select the needed data point for independant and dependant variable 

X = data.iloc[:,2].values.reshape(-1,1) #no.rows should be decided based on no.elements inside that array 
y = data.iloc[:,3].values.reshape(-1,1)
```

```python
y
```

x has to be a capital X  why?

```python
## linear regression model 
LR = LinearRegression()
#this is api call 
LR.fit(X,y)
```

```python
# we are making predictions on trained data 

y_pred = LR.predict(X)
y_pred #prediction of selling price of the car
```
