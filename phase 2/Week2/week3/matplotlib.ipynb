```python
import numpy as np
import matplotlib.pyplot as plt

y = np.linspace(-1, 1, 5)
x = np.arange(5)
fig = plt.figure()
ax = fig.subplots()
line1 = ax.plot(x, y)

# Plotting one more line on the same graph
y2 = np.linspace(-2, 2, 5)  # Numbers are random
line2 = ax.plot(x, y2)

# Display the plot
plt.show()
```

```python
x2 = np.arange(0, 4*np.pi, 0.05)
ycos = np.cos(x2)
ysin = np.sin(x2)

fig, ax =plt.subplots()
ax.plot(x2, ycos, 'b-', label = 'cos(x)')
ax.plot(x2, ysin, 'r--', label = 'sin(x)')
ax.set_title('Trigometric fuction')
ax.set_xlabel('0 to 4pi')
ax.set_ylabel('cos (x) and sin (x)')
ax.legend()


# b- for blue
# r - - for red
```
