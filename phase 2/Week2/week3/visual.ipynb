```python

# Select the top 5 earthquakes based on the 'mag' column
top_5_earthquakes = Earthquake.nlargest(5, 'mag')

# Plot all earthquakes
plt.scatter(Earthquake['longitude'], Earthquake['latitude'], label='All Earthquakes', color= 'pink', alpha=0.5)

# Highlight the top 5 earthquakes with a different color or marker
plt.scatter(top_5_earthquakes['longitude'], top_5_earthquakes['latitude'], color='black', label='Top 5 Earthquakes(based on magnitude)', marker='x')

# Set labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earthquakes - All and Top 5 Highlighted')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert 'time' to datetime format wtih to_datime function
Earthquake['time'] = pd.to_datetime(Earthquake['time'])

# Group by date and count the number of earthquakes
daily_counts = Earthquake.groupby(Earthquake['time'].dt.date).size().reset_index(name='total')

# Plotting the time series
plt.figure(figsize=(12, 8))
plt.plot(daily_counts['time'], daily_counts['total'], marker='o', linestyle='-', color='hotpink')
plt.title('Total Earthquakes Over Time')
plt.xlabel('Date')
plt.ylabel('Total Earthquakes')
plt.show()
```
